```python
import urllib.request

# URL yang akan diakses
url = 'https://www.traveloka.com/en-id/hotel/search?spec=05-10-2024.06-10-2024.1.1.HOTEL_GEO.102813.Jakarta.2&accommodationTypes=APARTMENT'

# Menyiapkan header
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:106.0) Gecko/20100101 Firefox/106.0',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.5',
}

# Membuat permintaan dengan header
req = urllib.request.Request(url, headers=headers)

# Melakukan request
with urllib.request.urlopen(req) as response:
    content = response.read()  # Mengambil konten sebagai bytes

# Menyimpan konten ke file HTML
with open("test.html", 'wb') as f:
    f.write(content)

# Membaca konten dari file HTML dan menyimpannya ke dalam response.content
with open("test.html", 'rb') as f:
    response_content = f.read()

# # Menampilkan hasil konten
print(response_content.decode('utf-8'))  # Dekode untuk menampilkan dalam format string
```

# Cleaning

```python
import re

# Fungsi untuk membersihkan harga (price)
def clean_price(price_str):
    # Pastikan price_str adalah string
    if isinstance(price_str, dict):
        # Jika price_str adalah dictionary, ambil field yang sesuai
        price_str = price_str.get('price', '')  # Ganti 'price' dengan key yang sesuai dari dict
    
    # Lakukan pembersihan jika price_str adalah string
    if isinstance(price_str, str):
        price_str = price_str.replace('Rp', '').replace('.', '').strip()  # Hapus 'Rp' dan titik
        return int(price_str)  # Konversi ke integer
    else:
        return 0  # Jika bukan string, kembalikan 0 atau nilai default lainnya

# Fungsi untuk memisahkan rating dan jumlah ulasan
def clean_rating(rating_str):
    # Pastikan rating_str adalah string
    if isinstance(rating_str, dict):
        # Jika rating_str adalah dictionary, ambil field yang sesuai
        rating_str = rating_str.get('rating', '')  # Ganti 'rating' dengan key yang sesuai dari dict
    
    # Lakukan pembersihan jika rating_str adalah string
    if isinstance(rating_str, str):
        # Ekstrak nilai rating
        rating_value = float(re.search(r'(\d+\.\d+)', rating_str).group(1))
        
        # Ekstrak jumlah ulasan dan konversi 'K' menjadi ribuan jika ada
        reviews = re.search(r'\(([\d\.]+[K]*) reviews\)', rating_str).group(1)
        if 'K' in reviews:
            reviews = float(reviews.replace('K', '')) * 1000
        else:
            reviews = float(reviews)
        
        return rating_value, int(reviews)
    else:
        return 0, 0  # Kembalikan nilai default jika bukan string
```

# Merge

```python
# Fungsi untuk merge data dan sekaligus melakukan cleaning
def merge_data(apartment_name, apartment_location, apartment_price, apartment_rating, apartment_features):
    all_data = []
    
    # Proses merge data
    for i in range(len(apartment_name)):
        # Pembersihan price
        cleaned_price = clean_price(apartment_price[i])
        
        # Pembersihan rating dan jumlah ulasan
        rating_value, total_reviews = clean_rating(apartment_rating[i])
        
        # Ambil nilai string dari dictionary (ambil nilai yang tepat dari setiap field)
        name_value = apartment_name[i].get('name', '') if isinstance(apartment_name[i], dict) else apartment_name[i]
        location_value = apartment_location[i].get('location', '') if isinstance(apartment_location[i], dict) else apartment_location[i]
        features_value = apartment_features[i].get('features', '') if isinstance(apartment_features[i], dict) else apartment_features[i]
        
        # Menggabungkan data yang sudah dibersihkan
        all_data.append({
            'name': name_value,  # Ambil nilai string dari dictionary
            'location': location_value,  # Ambil nilai string dari dictionary
            'price': cleaned_price,
            'rating': rating_value,
            'total_reviews': total_reviews,
            'features': features_value  # Ambil nilai string dari dictionary
        })
    
    return all_data

# Panggil fungsi merge 
all_data = merge_data(apartment_name, apartment_location, apartment_price, apartment_rating, apartment_features)

# Convert to DataFrame
import pandas as pd
df = pd.DataFrame(all_data)


# Lihat hasil
df
```

# Load to PostgreSQL

```python
from sqlalchemy import create_engine

# Membuat koneksi ke PostgreSQL
engine = create_engine('postgresql+psycopg2://postgres:admin@localhost:5432/Dibimbing')

# Memasukkan DataFrame ke dalam tabel apartment_data di PostgreSQL
df.to_sql('apartment_data', engine, if_exists='replace', index=False)

print("Data berhasil dimasukkan ke tabel apartment_data.")
```
