````python
# Import library
import pandas as pd
pd.set_option("display.max_columns",None)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Load Data
df = pd.read_csv("/content/Telco_customer_churn.csv")
df.head()
````

````python
df.info()
````

````python
# Standarisasi format data
# Mengubah Total Charge ke numerik
df['Total Charges'] = pd.to_numeric(df['Total Charges'], errors='coerce')
````

## Soal 1
- Tujuan dari visualisasi ini adalah untuk menunjukkan proporsi antara pelanggan yang churn (meninggalkan perusahaan) dan pelanggan yang tidak churn (tetap berlangganan).
- Pie chart adalah cara yang efektif untuk membandingkan proporsi atau persentase dari suatu kategori dalam total data. Karena data churn dan tidak churn hanya terdiri dari dua kategori, pie chart dapat memberikan visualisasi yang jelas tentang bagaimana kedua kelompok ini terdistribusi secara proporsional.

````python
churn_counts = df['Churn Label'].value_counts(normalize=True)

plt.figure(figsize=(6, 6))
plt.pie(churn_counts, labels=['Tidak Churn', 'Churn'], autopct='%1.1f%%', startangle=90, colors=['green', 'red'])
plt.title('Persentase Customer yang Churn dan Tidak Churn')
plt.show()
````

Insight :
- Persentase churn di atas 20% yakni 26,5% (sekitar seperempat dari jumlah keseluruhan pengguna) menunjukkan bahwa ada tantangan dalam mempertahankan pelanggan. Jika perusahaan tidak mengambil tindakan untuk mengurangi churn, mereka dapat kehilangan banyak pelanggan dan mengalami penurunan pendapatan jangka panjang.
- Perusahaan perlu menerapkan strategi retensi pelanggan yang lebih baik untuk mengurangi churn, seperti meningkatkan pengalaman onboarding, menawarkan insentif, dan memberikan layanan yang lebih responsif. Memastikan bahwa pelanggan mendapatkan nilai dari layanan dan merasa puas adalah kunci untuk meningkatkan retensi dan mengurangi churn dalam jangka panjang.

## Soal 2
- Tujuan dari visualisasi ini adalah untuk melihat hubungan atau korelasi antara tenure months (lama pelanggan berlangganan) dan total charges (total tagihan yang dibayarkan pelanggan).
- Scatter plot adalah pilihan terbaik untuk menunjukkan hubungan antara dua variabel numerik. Dalam kasus ini, scatter plot menampilkan tenure months di sumbu X dan total charges di sumbu Y. Dengan scatter plot, kita bisa langsung melihat apakah ada pola korelasi antara lama berlangganan dan total tagihan.

````python
plt.figure(figsize=(10, 6))
plt.scatter(df['Tenure Months'], df['Total Charges'], alpha=0.5)
plt.title('Hubungan Tenure Months dan Total Charges')
plt.xlabel('Tenure Months')
plt.ylabel('Total Charges')
plt.show()
````

Insight :
- Scatter plot diatas memperlihatkan hubungan linear antara dua variabel. Dari grafik, terlihat bahwa ada tren positif yakni semakin lama seorang pelanggan menjadi pelanggan perusahaan, semakin tinggi total tagihan mereka.
- Untuk setiap nilai Tenure Months, ada variasi yang cukup signifikan dalam Total Charges. Hal ini menunjukkan bahwa pelanggan dengan lama berlangganan yang sama bisa memiliki Total Charges yang berbeda, tergantung pada paket layanan yang mereka ambil atau tambahan layanan yang mereka gunakan.
- Sebagian besar pelanggan dengan Tenure Months rendah cenderung memiliki Total Charges yang rendah juga. Ini menandakan bahwa pelanggan baru mungkin belum melihat manfaat penuh dari layanan yang ditawarkan dan lebih rentan terhadap churn. Oleh karena itu, penting untuk memberikan pengalaman awal yang baik untuk mendorong mereka tetap berlangganan dan meningkatkan keterlibatan.

## Soal 3
- Tujuan dari visualisasi ini adalah untuk mengetahui kota-kota teratas di mana pelanggan cenderung churn atau tetap berlangganan.
- Stacked bar chart merupakan salah satu pilihan terbaik untuk membandingkan kategori dalam kelompok, dalam hal ini adalah churn dan tidak churn berdasarkan kota. Ini karena stacked bar chart dapat memberikan visualisasi yang jelas tentang jumlah pelanggan churn dan tidak churn di kota yang sama. Ini membantu kita melihat perbedaan antara kedua kategori (churn dan tidak churn) di kota yang berbeda dengan cepat.

````python
churn_by_city = df.groupby(['City', 'Churn Label']).size().unstack().fillna(0)
top_5_cities = churn_by_city.sum(axis=1).nlargest(5).index
top5_churn_by_city = churn_by_city.loc[top_5_cities]

top5_churn_by_city.plot(kind='bar', stacked=True, figsize=(10, 6), color=['green', 'red'])
plt.title('5 Kota Teratas dengan Churn dan Tidak Churn')
plt.ylabel('Jumlah Pelanggan')
plt.show()
````

Insight :
- Dari stacked bar chart di atas, dapat diketahui bahwa Los Angeles memiliki jumlah pelanggan yang jauh lebih tinggi dibandingkan kota-kota lain, termasuk pelanggan yang churn. Hal ini menunjukkan bahwa, meskipun Los Angeles merupakan pasar besar untuk perusahaan, ada masalah dengan retensi pelanggan yang perlu segera diatasi.
- Selain Los Angeles, empat kota lainnya yakni San Diego, San Jose, Sacramento, dan San Francisco juga menunjukkan jumlah pelanggan yang churn cukup signifikan, meskipun lebih sedikit dibandingkan Los Angeles.
- Perusahaan perlu fokus pada strategi yang spesifik untuk setiap kota berdasarkan kebutuhan dan masalah lokal. Strategi ini bisa berupa peningkatan kualitas layanan, promosi lokal, atau program loyalitas.

## Soal 4
- Tujuan dari visualisasi ini adalah untuk melihat bagaimana persebaran lama pelanggan berlangganan (tenure months) antara pelanggan yang churn dan tidak churn.
- Box plot adalah pilihan terbaik untuk menampilkan distribusi data serta outliers. Dalam konteks ini, kita ingin melihat distribusi tenure months untuk pelanggan yang churn dan tidak churn, dan apakah ada pola perbedaan antara keduanya.
- Box plot juga memungkinkan kita untuk membandingkan rentang (misalnya, median, kuartil pertama, kuartil ketiga) antara dua kelompok churn dan tidak churn. Selain itu, box plot dapat menyoroti outliers (pelanggan yang churn dengan lama berlangganan sangat tinggi atau sangat rendah), yang bisa menjadi wawasan penting untuk perusahaan.

````python
plt.figure(figsize=(10, 6))
df.boxplot(column='Tenure Months', by='Churn Label', grid=False)
plt.title('Persebaran Tenure Months Berdasarkan Churn Label')
plt.suptitle('')
plt.xlabel('Churn Label')
plt.ylabel('Tenure Months')
plt.show()
````

Insight :
- Dari box plot di atas dapat diketahui bahwa Pelanggan yang tidak churn memiliki tenure (lama berlangganan) yang lebih panjang secara keseluruhan dibandingkan dengan pelanggan yang churn.
- Median untuk pelanggan yang tidak churn adalah sekitar 35-40 bulan, menunjukkan bahwa pelanggan yang tetap setia biasanya memiliki waktu berlangganan yang lebih lama.
- Pelanggan yang churn cenderung memiliki tenure yang jauh lebih pendek, dengan median tenure sekitar 10 bulan. Ini menunjukkan bahwa pelanggan cenderung churn lebih cepat setelah bergabung dengan layanan.
- Interquartile range (IQR) untuk pelanggan churn lebih sempit (antara 0 hingga sekitar 25 bulan), yang berarti bahwa sebagian besar pelanggan yang churn meninggalkan layanan dalam waktu kurang dari 2 tahun.
- Pada pelanggan yang churn, terlihat ada outliers di bagian atas. Outliers ini menunjukkan bahwa ada beberapa pelanggan yang churn meskipun mereka telah berlangganan untuk waktu yang cukup lama (lebih dari 60 bulan).
- Dari visualisasi ini, jelas bahwa pelanggan yang churn umumnya adalah pelanggan baru yang tidak bertahan lama dengan layanan perusahaan. Ini berarti bahwa perusahaan mungkin memiliki masalah dalam retensi pelanggan baru, dan perlu meningkatkan pengalaman pelanggan di bulan-bulan awal layanan mereka untuk mengurangi churn.
