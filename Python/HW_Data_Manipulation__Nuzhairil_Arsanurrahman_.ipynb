# Soal 1
Buat fungsi Python untuk memfilter churn untuk menemukan pelanggan yang berisiko tinggi.
Pelanggan berisiko tinggi didefinisikan sebagai pelanggan yang memiliki kontrak bulanan
dan tagihan bulanan yang tinggi (di atas persentil 90%).

```python
#cek jenis kontrak
df['Contract'].unique()
```

```python
#Cek persentil 90%
print(f"Persentil 90% dari tagihan bulanan : {df['MonthlyCharges'].quantile(0.9)}")
```

```python
#Syarat yang harus terpenuhi (Kontraknya = bulanan dan tagihan > persentil 90% tagihan bulanan)
condition = (df['Contract']=='Month-to-month') & (df['MonthlyCharges']>df['MonthlyCharges'].quantile(0.9))
```

```python
#melakukan filtering sesuai syarat
df_copy1 = df[condition]
df_copy1
```

```python
df_copy1['MonthlyCharges'].describe()
```

```python
#Membuat function
def filter_churn(df):
  condition = (df['Contract']=='Month-to-month') & (df['MonthlyCharges']>df['MonthlyCharges'].quantile(0.9))
  return df[condition]
```

```python
#Memanggil function
filter_churn(df)
```

# Soal 2
Buat script Python untuk mengelompokkan data berdasarkan jenis kontrak dan menghitung
rata-rata tagihan bulanan serta persentase churn untuk setiap grup (1 dan 0, rata rata untuk
menghitung persentase)

```python
#Menghitung rata-rata tagihan bulanan dan churn rate
df[['MonthlyCharges','Churn']].mean()
```

```python
# Mengelompokkan data dengan group by berdasarkan kontrak
df_copy2 = df.groupby(['Contract'])[['MonthlyCharges','Churn']].mean()
df_copy2
```

```python
#Churn rate tiap grup
df_copy2['Churn'] = df_copy2['Churn']*100
df_copy2
```

# Soal 3
Buat pivot table untuk menganalisis hubungan antara jenis kontrak dan metode pembayaran
terhadap rata-rata tagihan bulanan.

```python
#Membuat pivot table yang menampilkan rata-rata tagihan dari setiap metode pembayaran pada setiap jenis kontrak
df_copy3 = pd.pivot_table(df, values='MonthlyCharges',index='Contract', columns='PaymentMethod', aggfunc='mean')
df_copy3
```

# Soal 4
Buat crosstab untuk mengevaluasi hubungan antara churn dan metode pembayaran, serta
menghitung persentase churn untuk setiap metode pembayaran.

```python
# Hubungan antara churn dan metode pembayaran
df_copy4 = pd.crosstab(df['Churn'], df['PaymentMethod'])
df_copy4
```
