### Problem 1 Sort Colors :
https://leetcode.com/problems/sort-colors/

- Step Jawaban : 5 poin

- Code : 10 poin

Langkah-langkah mengerjakan soal tanpa menggunakan fungsi sort :
1. Menentukan posisi dari tiap-tiap warna, yaitu:
- Warna merah (0), semuanya terletak dibagian awal array
- Warna putih (1), semuanya terletak dibagian tengah array
- Warna biru (2), semuanya terletak dibagian akhir array
2. Menggunakan indeks posisi untuk patokan melakukan swap(penukaran posisi antar elemen di array) dengan ketentuan
- bottom : untuk menunjukkan posisi awal array (untuk posisi warna merah (0)) dan juga sebagai batas bawahnya
- middle : untuk menunjukkan posisi tengah array sekaligus sebagai posisi yang akan berjalan dalam loop untuk melakukan pengecekkan
- top : untuk menunjukkan posisi akhir array (untuk posisi warna biru (2)) dan juga sebagai batas atasnya
3. Melakukan penukaran posisi dengan ketentuan
- Apabila posisi yang di cek saat ini sudah benar maka tidak akan terjadi penukaran dan posisi pengecekkan lanjut ke elemen berikutnya
- Apabila posisi yang di cek tidak sesuai maka akan dilakukan penukaran posisi dengan ketentuan :
  - Apabila posisi yang dicek saat ini adalah warna merah (0) maka akan dilakukan penukaran dengan elemen pada posisi batas bawah dan nilai batas bawahnya akan di increment
  - Apabila posisi yang dicek saat ini adalah warna biru (2) maka akan dilakukan penukaran dengan elemen pada posisi batas atas dan nilai batas atasnya akan di incdecrementrement


```python
# write your code here
def sortColors(nums):
    # Inisialisasi posisi bottom (0), middle, dan top (2)
    bottom, middle, top = 0, 0, len(nums) - 1

    # Loop hingga middle melewati top
    while middle <= top:
        if nums[middle] == 0:  # Jika 0, tukar ke posisi bottom dan geser kedua indeks
            nums[bottom], nums[middle] = nums[middle], nums[bottom]
            bottom += 1
            middle += 1
        elif nums[middle] == 1:  # Jika 1, lanjutkan ke elemen berikutnya
            middle += 1
        else:  # Jika 2, tukar ke posisi top dan geser top ke kiri
            nums[middle], nums[top] = nums[top], nums[middle]
            top -= 1
```

```python
# contoh penerapan:
nums1 = [2,0,2,1,1,0]
nums2 = [2,0,1]
sortColors(nums1)
sortColors(nums2)
print(nums1)
print(nums2)
```

## Problem 2 letter combinations of a phone number :
https://leetcode.com/problems/letter-combinations-of-a-phone-number/

- Step jawaban : 10

- Code : 15 Poin

Langkah-langkah dalam mengerjakan soal
1. Membuat dictionary untuk menyimpan angka serta huruf yang ada dalam tombol nomor di handphone
2. Membuat fungsi rekursif(backtracking) untuk mengumpulkan kombinasi huruf dimana dalam fungsi ini akan dilakukan pengambilan digit serta hurufnya dari dictionary sebelumnya dan akan dilakukan penggabungan kombinasi huruf antar digit sesuai dengan jumlah inputan

```python
def letterCombinations(digits):
    # Membuat dictionary
    digit_to_char = {
        "2": "abc", "3": "def", "4": "ghi", "5": "jkl",
        "6": "mno", "7": "pqrs", "8": "tuv", "9": "wxyz"
    }

    # Jika input kosong, kembalikan list kosong
    if not digits:
        return []

    # Inisialisasi hasil
    result = []

    # Fungsi rekursif untuk mengumpulkan kombinasi
    def backtrack(index, current_combination):
        if index == len(digits):
            result.append(current_combination)
            return

        # Dapatkan huruf yang sesuai dengan digit saat ini
        current_digit = digits[index]
        possible_chars = digit_to_char[current_digit]

        # Iterasi untuk setiap huruf dan lanjutkan rekursi
        for char in possible_chars:
            backtrack(index + 1, current_combination + char)

    # Memulai proses rekursi
    backtrack(0, "")

    # Mengembalikan hasil akhir
    return result

# Contoh penggunaan:
print(letterCombinations("23"))
```

### 1. Soal : Tiap Perusahaan ingin menghitung suatu pesanan pada tiap harinya yang memiliki status rejected.
#### (poin 15)
Saran pengerjaan :

Anda memiliki tabel orders dengan kolom-kolom berikut: company, order_date, status, dan order_id.

Tulislah sebuah query SQL yang menghasilkan:

1. Kolom company
2. Kolom tanggal yang merupakan hari dari tanggal order (dalam format tanggal dua digit).
3. Kolom total_rejected yang merupakan jumlah order yang memiliki status 'REJECTED' pada setiap perusahaan (company) pada setiap tanggal order.

Output dari query tersebut harus mengelompokkan hasil berdasarkan company dan tanggal order (tanggal), serta menghitung jumlah order yang memiliki status 'REJECTED' untuk setiap kelompok.

```python
query = """
SELECT
    *
FROM
    orders
"""

sql_run(query)
```

```python
query = """
SELECT
    company,
    strftime('%d', order_date) AS tanggal,
    COUNT(order_id) AS total_rejected
FROM
    orders
WHERE
    status = 'REJECTED'
GROUP BY
    company,
    strftime('%d', order_date)
ORDER BY
    company, tanggal;
"""

sql_run(query)
```

### 2. Atasan anda ingin mengetahui apa saja produk yang telah dipesan oleh setiap pelanggan (customer_id) bersama dengan jumlahnya (quantity) pada setiap pesanan?
#### (poin 15)
Saran pengerjaan :
Anda memiliki dua tabel, yaitu orders dan order_details, dengan kolom-kolom berikut:

 1.   Tabel orders memiliki kolom customer_id, order_id, dan mungkin kolom lain terkait pesanan.
 2.   Tabel order_details memiliki kolom order_id, product_name, quantity, dan mungkin kolom lain terkait detail produk.

Tuliskan sebuah query SQL yang melakukan hal berikut:

 1.   Memilih kolom customer_id, order_id, product_name, dan quantity dari kedua tabel tersebut.
 2.   Menggabungkan kedua tabel berdasarkan order_id.
 3.   Mengurutkan hasil berdasarkan customer_id dan order_id.

```python
query = """
SELECT
    *
FROM
    order_details
"""

sql_run(query)
```

```python
query = """
SELECT
    o.customer_id,
    o.order_id,
    od.product_name,
    od.quantity
FROM
    orders o
JOIN
    order_details od
ON
    o.order_id = od.order_id
ORDER BY
    o.customer_id,
    o.order_id;
"""

sql_run(query)
```

### 3. Seperti biasa, atasa anda ingin menampilkan total orders dan kategori di tiap perusahaan dengan beberapa kriteria berikut :
Saran Pengerjaan :
1. category harus :
2. Good (total order status sent > 2 dan total rejected > 0),
3. Superb (tidak ada yang di reject, dan total order status sent > 2),
4. selain ini dia Normal

### notes :
1. Kerjakan secara CTE **(poin 15)**
2. kerjakan secara Nested Query **(poin 15)**
3. Harusnya hasilnya sama, Apakah hasilnya sama setelah anda running jawaban pada poin 1 dan 2 diatas ?

```python
# Dengan CTE
query = """
WITH order_summary AS (
    SELECT
        o.company AS company,
        SUM(CASE WHEN o.status = 'SENT' THEN 1 ELSE 0 END) AS total_sent,
        SUM(CASE WHEN o.status = 'REJECTED' THEN 1 ELSE 0 END) AS total_rejected
    FROM
        orders o
    JOIN
        order_details od ON o.order_id = od.order_id
    GROUP BY
        o.company
),
categorized_orders AS (
    SELECT
        company,
        total_sent,
        total_rejected,
        CASE
            WHEN total_sent > 2 AND total_rejected > 0 THEN 'Good'
            WHEN total_sent > 2 AND total_rejected = 0 THEN 'Superb'
            ELSE 'Normal'
        END AS category
    FROM
        order_summary
)
SELECT * FROM categorized_orders;

"""

sql_run(query)
```

```python
#Dengan Sub Query
query = """
SELECT
    company,
    total_sent,
    total_rejected,
    CASE
        WHEN total_sent > 2 AND total_rejected > 0 THEN 'Good'
        WHEN total_sent > 2 AND total_rejected = 0 THEN 'Superb'
        ELSE 'Normal'
    END AS category
FROM (
    SELECT
        o.company AS company,
        SUM(CASE WHEN o.status = 'SENT' THEN 1 ELSE 0 END) AS total_sent,
        SUM(CASE WHEN o.status = 'REJECTED' THEN 1 ELSE 0 END) AS total_rejected
    FROM
        orders o
    JOIN
        order_details od ON o.order_id = od.order_id
    GROUP BY
        o.company
) AS order_summary;
"""

sql_run(query)
```
